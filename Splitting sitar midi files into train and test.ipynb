```python
import pandas as pd
import os
```

```python
base_path = '/home/nisarg/inhouse_research/imusic/third_party_repos/repo_data/'
# train_path = os.path.join(base_path, 'maestro_sitar_train')
# test_path = os.path.join(base_path, 'maestro_sitar_test')
parent_folder = os.path.join(base_path, 'sitar_lmd_files')
```

```python
all_files = os.listdir(parent_folder)
train_files = all_files[:int(0.8*len(all_files))]
test_files = all_files[int(0.8*len(all_files)):]
```

```python
df = pd.DataFrame()
rows = []
```

```python
for train_file_wav in train_files:
    if not train_file_wav.endswith('wav'):
        continue
    train_file_wav = os.path.join(parent_folder, train_file_wav)
    train_file_midi = train_file_wav.replace('.wav', '.midi')
    row = {}
    row['audio_filename'] = train_file_wav
    row['midi_filename'] = train_file_midi
    row['split'] = 'train'
    rows.append(row)
```

```python
for test_file_wav in test_files:
    if not test_file_wav.endswith('wav'):
        continue
    test_file_wav = os.path.join(parent_folder, test_file_wav)
    test_file_midi = test_file_wav.replace('.wav', '.midi')
    row = {}
    row['audio_filename'] = test_file_wav
    row['midi_filename'] = test_file_midi
    row['split'] = 'test'
    rows.append(row)
```

```python
df = pd.DataFrame(rows)
```
